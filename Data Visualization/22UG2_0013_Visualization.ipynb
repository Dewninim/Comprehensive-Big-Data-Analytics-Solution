```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "processed_healthcare_data (1).csv"
df = pd.read_csv(file_path)
```

```python
# Basic data overview
print(df.info())
print(df.head())
```

```python
# Convert date columns to datetime format
df['date_of_admission'] = pd.to_datetime(df['date_of_admission'])
df['discharge_date'] = pd.to_datetime(df['discharge_date'])
```

```python
# Visualization 1: Distribution of Age
df['age'].hist(bins=30, edgecolor='black')
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Distribution of Age among Patients")
plt.show()
```

```python
# Visualization 2: Gender Distribution
sns.countplot(x='gender', data=df, hue='gender', palette='coolwarm', legend=False)
plt.title("Gender Distribution of Patients")
plt.show()
```

```python
# Visualization 3: Admission Type Count
sns.countplot(y='admission_type', data=df,
              order=df['admission_type'].value_counts().index,
              palette='viridis', hue='admission_type', legend=False)
plt.title("Types of Admissions")
plt.show()
```

```python
# Visualization 4: Billing Amount Distribution
sns.histplot(df['billing_amount'], bins=30, kde=True, color='blue')
plt.xlabel("Billing Amount ($)")
plt.ylabel("Frequency")
plt.title("Distribution of Billing Amounts")
plt.show()
```

```python
# Visualization 5: Medical Condition Count
df['medical_condition'].value_counts().head(10).plot(kind='bar', color='green', edgecolor='black')
plt.xlabel("Medical Condition")
plt.ylabel("Count")
plt.title("Top 10 Medical Conditions")
plt.xticks(rotation=45)
plt.show()
```
